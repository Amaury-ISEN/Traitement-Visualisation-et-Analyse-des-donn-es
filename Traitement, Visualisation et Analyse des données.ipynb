### 5.3 Analyse de corrélations
L’objectif de l’analyse de corrélation est étudier la liaison entre deux variables quantitatives.

Soient X et Y deux grandeurs statistiques quantitatives observées. On souhaite :

• Déterminer s’il existe une relation entre X et Y .

• Caractériser la forme de la liaison(la relation) entre X et Y (positive ou négative,linéaire ou non linéaire,monotone ou non monotone)

• Tester si la liaison est statistiquement significative

• Quantifier l’intensité de la liaison

```python
df = pd.read_csv("./Data_Vis/iris.csv")
df
```

On peut dresser un tableau des corrélations (de Pearson, par défaut) avec la méthode corr() de pandas :

```python
df.corr()
```

Il est aussi possible d'utiliser stats de scipy pour calculer la corrélation de Pearson pour deux variables :

```python
x = df["longueur_petal"]
y = df["largeur_petal"]

coeff_p, p_val = sp.stats.pearsonr(x , y)
print("corrélation de Pearson :", coeff_p)
print("p_value :", p_val)

p_val < 0.05
```

Cette seconde méthode renvoie le coeff de Pearson mais aussi la p-value, c'est à dire la probabilité qu'un système non-corrélé produise un dataset ayant une corrélation de Pearson au moins aussi élevée que celle calculée.

Cette p-value est surtout pertinente pour de "gros" datasets (+500 lignes).

A priori, plus la p-value est faible et moins la corrélation observée peut être dûe au hasard.

```python
#Transformation des data series pandas en ndarray utilisables par l'algo sklearn :
x = x.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)

# Applying linear reg
from sklearn.linear_model import LinearRegression
regr = LinearRegression().fit(x,y)

# Calcul du R2
from sklearn.metrics import r2_score
print(r2_score(regr.predict(x), y))
```

Je considère qu'un R2 de 92% pour une régression linéaire sur notre dataset "longueur pétal" par rapport à "largueur pétal" signifie que la relation qui lie ces deux variables est bien linéaire et positive.

#### 5.3.2 Analyse mathématique

Coefficient de corrélation de Pearson Le coefficient de corrélation linéaire simple, dit de Bravais-Pearson(ou de Pearson), est une normalisation de la covariance par le produit des écarts-type des variables : rx,y = COV(X,Y) / (E(X) x E(Y)) Avec les outils et bibliothèques de Python, calculez le coefficient de corrélation de Pearson sur la base de données fournie et interprétez les résultats.

```python
df.corr()
```

**Réponse :**

La question a partiellement été traitée plus haut. On peut rajouter que le coefficient de Pearson va de -1 à 1.

**Pour un coeff très proche de zéro :**

cela signifie que les variables ne sont pas du tout ou très peu corrélées.

**Pour un coeff qui s'approche de 1 :**

forte corrélation positive (c'est-à-dire que les deux variables augmentent ensemble).

**Pour un coeff qui s'approche de -1 :**

forte corrélation inverse (c'est-à-dire que lorsqu'une variable augmente, l'autre a tendance à diminuer de façon liée).

Une variable comparée à elle-même donne évidemment un coefficient de 1 (corrélation parfaite).


**Notre corrélation est-elle statistiquement significative ?**

En alliant la *p-value* au coeff de corrélation on peut voir si une corrélation est statistiquement significative.
En général on prend le seuil de 0.05 pour la *p-value*.

Dans le cadre de la méthode pearsonr() du module stats de Scipy, si on est au-dessous, cela signifie que la corrélation est statistiquement significative (mais sans préjuger d'une éventuelle causalité).
